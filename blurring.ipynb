```python
import cv2

from google.colab.patches import cv2_imshow
img = cv2.imread('/content/horse.jpeg')

cv2_imshow(img)
```

```python
# avarage blur

avg_blur = cv2.blur(img, (5,5)) # func. gets the destination and kernel size

cv2_imshow(avg_blur)
```

```python
# gaussian Blur
ga_blur = cv2.GaussianBlur(img, (5,5), 0)

cv2_imshow(ga_blur)
```

Math in the back of Median Blurring : https://www.youtube.com/watch?v=eJx3g-ZEfm4

```python
#median blurring

median_blur = cv2.medianBlur(img, 5)

cv2_imshow(median_blur)
```

This type of blurring is much more better then others but when I applied this techn., I have a problem with edges so we have to use other type of blurring which is `'bilateral blurring'`.

```python
# bilateral Blur

bilateral = cv2.bilateralFilter(img, 5, 15,15)

cv2_imshow(bilateral)
```
