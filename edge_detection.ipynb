Previous I used Canny to edge detection, this time 2 more functions use to edge detections: https://pyimagesearch.com/2021/05/12/opencv-edge-detection-cv2-canny/

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/horse.jpeg')

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
# Laplacion to detect the edge

import numpy as np
lap_img = cv2.Laplacian(gray_img, cv2.CV_64F)
lap = np.uint(np.absolute(lap_img)) # image can't take negative pixel, values so i used absolute to convert all pixels to positive
cv2_imshow(lap) 
```

```python
# Sobel to detect the edge

sobelX = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0) # direction of X

sobelY = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1) # direction of Y

cv2_imshow(sobelX)

cv2_imshow(sobelY)
```

```python
# combine the Sobal X and Y

combine_sobal = cv2.bitwise_or(sobelX, sobelY)
cv2_imshow(combine_sobal)
```

```python
# CANNY vs SOBEL

canny_img = cv2.Canny(gray_img,150, 175)

cv2_imshow(canny_img)
```
