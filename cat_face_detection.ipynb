```python
# import libraries

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# import image

cat_img = cv2.imread('/content/fake_cats.tiff')

cv2_imshow(cat_img)
```

```python
# BGR to Grayscale

cat_img_grayscale = cv2.cvtColor(cat_img, cv2.COLOR_BGR2GRAY)

cv2_imshow(cat_img_grayscale)
```

```python
# read haar cascade file

haar_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalcatface_extended.xml')
```

```python
# get the image and use cascade instance to find it
cat_face_rect = haar_cascade.detectMultiScale(cat_img_grayscale, scaleFactor = 1.2, minNeighbors = 3)
```

```python
# check how many cat faces are found

print(f'There is {len(cat_face_rect)} cat face in the given image.')
```

```python
for (x, y, w, h) in cat_face_rect:
    cv2.rectangle(cat_img, (x,y), (x+w,y+h), (0,255,0), thickness = 2)

cv2_imshow(cat_img)
```
