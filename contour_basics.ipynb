# **What are Contours?** 

When we join all the points on the boundary of an object, we get a contour. Typically, a specific contour refers to boundary pixels that have the same color and intensity. OpenCV makes it really easy to find and draw contours in images. It provides two simple functions:
`findContours()` `drawContours()`

Also, it has two different algorithms for contour detection: `CHAIN_APPROX_SIMPLE` `CHAIN_APPROX_NONE`





**Importing libraries :**

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

**Importing Image and change it to Grayscale:**

```python
# import image 
img = cv2.imread('/content/shapes.png')

# change it to grayscale
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img_grayscale)
```

```python
canny = cv2.Canny(img_grayscale, 125, 175)
cv2_imshow(canny)
```

```python
# applying threshold

ret, thresh = cv2.threshold(img_grayscale, 125, 255, cv2.THRESH_BINARY)

cv2_imshow(thresh)
```


**OpenCV: Contours Hierarchy**
This way, contours in an image has some relationship to each other. And we can specify how one contour is connected to each other, like, is it child of some other contour, or is it a parent etc. Representation of this relationship is called the Hierarchy.

```python
# Finding all contours in image
contours, hierarchies = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Display the number of contours
print(f'There are total {len(contours)} in the image.')
```

Retrieval Modes:

`cv2.RETR_LIST ` : takes everything in the given image.

`cv2.RETR_EXTERNAL` : takes all external layers of the image

`cv2.RETR_TREE ` : takes like parent and child contour

`'cv2.RETR_CCOMP` : takes levels seperatly and work on it.

```python
# drawing to the image by using thresh

cv2.drawContours(img, contours, -1, (0,255,0), 2)

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
