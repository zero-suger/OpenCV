```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# import image

team_img = cv2.imread('/content/chelasea_team.jpeg')

cv2_imshow(team_img)
```

```python
team_grayscale_img = cv2.cvtColor(team_img, cv2.COLOR_BGR2GRAY)
cv2_imshow(team_grayscale_img)
```

```python
haar_cascade = cv2.CascadeClassifier('/content/haarcascade_profileface.xml')
```

```python
team_face_rec = haar_cascade.detectMultiScale(team_grayscale_img, scaleFactor = 1.1, minNeighbors = 1)
print(f'There are {len(team_face_rec)} human faces in the given image.')
```

```python
for (x , y, w, h) in team_face_rec:
  cv2.rectangle(team_img, (x,y), (x+w,y+h), (0,255,0), thickness = 2)

cv2_imshow(team_img)
```
