```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/horse.jpeg')


cv2_imshow(img)
```

# **SIMPLE THRESHOLDING**

```python
# simple thresholding

# convert it into grayscale image

grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 1st simple thresh

threshold, thresh = cv2.threshold(grayscale_img, 100, 255, cv2.THRESH_BINARY) # set value is threshold, in my case it is '150'
cv2_imshow(thresh)
```

```python
# 2nd inversing the thresh

grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

threshold, thresh_inv = cv2.threshold(grayscale_img, 100, 255, cv2.THRESH_BINARY_INV) # set value is threshold, in my case it is '150'
cv2_imshow(thresh_inv) # this is what inversing  
```

# **ADOPTIVE THRESHOLDING**
https://pyimagesearch.com/2021/05/12/adaptive-thresholding-with-opencv-cv2-adaptivethreshold/

```python
# simple thresholding

adoptive_thresh = cv2.adaptiveThreshold(grayscale_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 3)

cv2_imshow(adoptive_thresh)
```

```python
# inv. thresholding

adoptive_thresh = cv2.adaptiveThreshold(grayscale_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 3)

cv2_imshow(adoptive_thresh)
```
