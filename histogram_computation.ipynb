**Histogram Computation :** A histogram represents the distribution of pixel intensities (whether color or grayscale) in an image. It can be visualized as a graph (or plot) that gives a high-level intuition of the intensity (pixel value) distribution. We are going to assume a RGB color space in this example, so these pixel values will be in the range of 0 to 255. (https://pyimagesearch.com/2021/04/28/opencv-image-histograms-cv2-calchist/)

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/horse.jpeg')

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray_img)

```

calHist takes all in. as list: 1. (where is the image?), 2.color chanel (BGR is 1 and GrayScale is 0, 3. masked part , 4.bins, 5. range )

```python
# grayscale image histogram

gray_histogram = cv2.calcHist([gray_img], [0], None, [256], [0,256]) 

plt.figure()
plt.title('My uploaded image histogram.')
plt.xlabel('Bins')
plt.ylabel('* of pixels')
plt.plot(gray_histogram)
plt.xlim([0,256])
plt.show()

cv2.waitKey(0)
```

```python
# BGR image histogram

colors = ('b', 'g', 'r')

for i, color_img in enumerate(colors):
  histogram_img = cv2.calcHist([img], [i], None, [256],[0,256])

  plt.plot(histogram_img, color = color_img)
  plt.xlim([0,256])

plt.show()
```
