**Splitting colors to R G B color channels.**

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/horse.jpeg')

cv2_imshow(img)
```

```python
B, G, R = cv2.split(img) # spliting the image into BGR form
cv2_imshow(B)
cv2_imshow(G)
cv2_imshow(R)
```

```python
# merging back images

img_merge = cv2.merge([B, G, R]) # use the merge function to merge back my images

cv2_imshow(img_merge)
```
