# **Drawing and putting text in OpenCV**

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

img = cv.imread('/cat.jpeg')



def resizeImage(frame, scale = 0.15):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)

    dimension = (width, height)

    return cv.resize(frame, dimension, interpolation = cv.INTER_AREA)

#cv.putText(img, "Hello, I am cat.", (0,0), cv.FRONT_HARSHEY_TRIPLEX, 1.0, (0,0,255),3) # add a text

new_size = resizeImage(img)




cv2_imshow(new_size)

```

# **Print the simple shape 500x500**

```python
blank = np.zeros((500,500,3), dtype = 'uint8')
cv2_imshow(blank)
```

```python
blank[:] = 0,0,255 # take all width and height
cv2_imshow(blank)
```

```python
blank[200:300, 300:400] = 0,255,0 
cv2_imshow(blank)
```

# **Drawing a rectangular**

```python
cv.rectangle(blank, (0,0),(250,250), (255,0,0), thickness = 2)  # gets the place to draw, get starting point, ending point, thikness, 
cv2_imshow(blank)
```

```python
cv.rectangle(blank, (0,0),(250,250), (255,0,0), thickness = cv.FILLED,)  # gets the place to draw, get starting point, ending point, thikness can get number and also cv.Filled to fill the place
cv2_imshow(blank)
```

```python
starting_point = (0,0)
ending_point = (250,500)
cv.rectangle(blank, starting_point, ending_point, (0,255,0))#thickness = -1,)  # gets the place to draw, get starting point, ending point, thikness can get number and also cv.Filled to fill the place or -1
cv2_imshow(blank)
```

# **Drawing a circle**

```python
cv.circle(blank, (150, 150), 30, (0,255,255), cv.FILLED) # gets : where?, starting_point, ending point, radius, color, extra parameters

cv2_imshow(blank)
```

# **Drawing a Line**

```python
cv.line(blank, (0,0),(500,500), (255,255,0), thickness = 5,)  
cv2_imshow(blank)
```
