# **Masking** is basically focusing on the part of the image which we want to work on.

https://pyimagesearch.com/2021/01/19/image-masking-with-opencv/

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img = cv2.imread('/content/horse.jpeg')

blank = np.zeros(img.shape[:2], dtype = 'uint8') # the shape of blank MUST be the same as my image shape
cv2_imshow(blank)
```

```python
# I want to draw circle and call it as the same name with musk

musk_circle = cv2.circle(blank,(img.shape[1]//2 , img.shape[0] // 2), 100, 255, -1) # the circle I want to draw
masked_img = cv2.bitwise_and(img,img, mask = musk_circle)

cv2_imshow(masked_img)
```
