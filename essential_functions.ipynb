# **`The most 6 essential functions in OpenCV`**

# **1st Converting the RGB to Grayscale:**

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

img = cv.imread('/content/cat.jpeg')

convert_grayscale = cv.cvtColor(img, cv.COLOR_BGR2GRAY) 


def resizeImage(frame, scale = 0.15):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    

    dimension = (width, height)

    return cv.resize(frame, dimension, interpolation = cv.INTER_AREA)



new_image = resizeImage(convert_grayscale)

cv2_imshow(new_image)


```

# **2nd Bluring Image :**

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

img = cv.imread('/content/cat.jpeg')


blur = cv.GaussianBlur(img,(7,7), cv.BORDER_DEFAULT)

def resizeImage(frame, scale = 0.15):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    

    dimension = (width, height)

    return cv.resize(frame, dimension, interpolation = cv.INTER_AREA)



new_image = resizeImage(blur)

cv2_imshow(new_image)
```

# **3rd Edge Cascade :**

```python
canny = cv.Canny(img,150,180)

def resizeImage(frame, scale = 0.15):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    

    dimension = (width, height)

    return cv.resize(frame, dimension, interpolation = cv.INTER_AREA)



new_image = resizeImage(canny)

cv2_imshow(new_image)
```

# **4th Dilating the Image**

```python
canny = cv.Canny(img,150,180)
dilated = cv.dilate(canny,(7,7), iterations = 1)

new_image = resizeImage(dilated)

cv2_imshow(new_image)
```

# **5th Resize :**

```python
resized = cv.resize(img,(500,500))
cv2_imshow(resized)
```
